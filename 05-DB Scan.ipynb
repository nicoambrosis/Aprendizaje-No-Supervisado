```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn import cluster, datasets, mixture
from sklearn.neighbors import kneighbors_graph
from sklearn.preprocessing import StandardScaler
from itertools import cycle, islice
from sklearn.cluster import DBSCAN

np.random.seed(0)
```

```python
n_samples = 1500
random_state = 170
```

## DataSet #1

```python
#circulos concentricos con ruido (tipo diana para tirar dardos)
noisy_circles, noisy_circles_labels = datasets.make_circles(n_samples=n_samples,
                                                            factor=.5,
                                                            noise=0.05,
                                                            random_state=random_state
                                                            )

# graficamos los datos generados
plt.figure(figsize=(6,6))
plt.plot(noisy_circles[: , 0],noisy_circles[: , 1],
         'o',
         markerfacecolor='lightBlue',
         markeredgecolor='k', markersize=10)
plt.xlabel('variable 1')
plt.ylabel('variable 2')
plt.show()
```

```python
dbscan = DBSCAN(eps=.2)
dbscan.fit_predict(noisy_circles)
```

```python
np.unique(dbscan.fit_predict(noisy_circles))
```

```python
plt.figure(figsize=(8,5))
plt.scatter(noisy_circles[: , 0],noisy_circles[: , 1],
            edgecolors = 'k',
            s=75,
            c = dbscan.fit_predict(noisy_circles),
            cmap='Set3'
            )
plt.xlabel('variable 1')
plt.ylabel('variable 2')
plt.show()
```

## DataSet #3

```python
# circulos con distintas varianzas por lo cual estan cerquita y pueden tener tambien diferente tamao de muestra
varied, varied_labels = datasets.make_blobs(n_samples=n_samples,
                             cluster_std=[1.0, 2.5, 0.5],
                             random_state=random_state)
# graficamos los datos generados
plt.figure(figsize=(6,6))
plt.plot(varied[: , 0],varied[: , 1],
         'o',
         markerfacecolor='#3296ba',
         markeredgecolor='k', markersize=10)
plt.xlabel('variable 1')
plt.ylabel('variable 2')
plt.show()
```

```python
dbscan = DBSCAN()
dbscan.fit_predict(varied)
```

```python
np.unique(dbscan.fit_predict(varied))
```

```python
color = dbscan.fit_predict(varied)
plt.figure(figsize=(8,5))
plt.scatter(varied[: , 0],varied[: , 1],
            edgecolors = 'k',
            s=75,
            c = color,
            cmap='Set3'
            )
plt.xlabel('variable 1')
plt.ylabel('variable 2')
plt.legend()
plt.show()
```
